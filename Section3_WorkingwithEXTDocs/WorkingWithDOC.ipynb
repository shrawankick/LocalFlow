```python

# Ensure native HEIF/HEIC support is installed before importing loaders
%pip install --upgrade pillow-heif pyheif pylibheif

import os
# Try importing the loader; if a missing native extension error occurs, install pillow-heif and retry.
try:
    from langchain_community.document_loaders import UnstructuredPDFLoader
except ModuleNotFoundError as e:
    # If the error is about a missing heif native module (e.g. 'pi_heif'), ensure pillow-heif is installed and retry.
    # Re-run the installation and import again in the same cell.
    print("ModuleNotFoundError during import:", e)
    %pip install --upgrade pillow-heif
    from langchain_community.document_loaders import UnstructuredPDFLoader

docs_folder = "./Docs"
documents = []

for filename in os.listdir(docs_folder):
    filepath = os.path.join(docs_folder, filename)
    loader = UnstructuredPDFLoader(filepath)
    documents.extend(loader.load())
```
